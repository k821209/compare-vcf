```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot') 
```

```python
import matplotlib.cm as cm
n=3 # number of colors
colors=cm.rainbow(np.linspace(0,1,n))
```

```python
df = pd.read_csv('pepper.allgenes.snp.pos.txt.gene2snp.txt',sep='\t')
```

```python
plt.figure(1,figsize=(10,10))

plt.hist(df['num_cds_snp'],bins=np.arange(0,100,1),label='CDS SNP',alpha=1)
plt.hist(df['num_nonsyn_snp'],bins=np.arange(0,100,1),label='nonsyn SNP',alpha=1)
plt.hist(df['num_genic_snp'],bins=np.arange(0,100,1),label='Genic SNP',alpha=1)
plt.xlim(0,20)
plt.legend()
```

```python
fig, ax = plt.subplots(1,sharex=True,sharey=True,figsize=(10, 7))
fig1, ax2 = plt.subplots(1,sharex=True,sharey=True,figsize=(10, 7))
inlist = [df['num_genic_snp'],df['num_cds_snp'],df['num_nonsyn_snp']]
labels = ['num_genic_snp','num_cds_snp','num_nonsyn_snp']
x = np.arange(0,100,2) #for CDS
for i,j in enumerate(inlist):
    a,b,c = ax.hist(j,bins=x,color=colors[i],label=labels[i])
    ax2.plot(b[0:len(a)],a,alpha=0.9,color=colors[i],label=labels[i])
plt.legend()
```
